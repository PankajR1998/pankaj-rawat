```python
from dotenv import load_dotenv
import os
from langchain.llms import HuggingFaceHub
from langchain.chains.llm import LLMChain
load_dotenv()
model = HuggingFaceHub(repo_id='mistralai/Mistral-7B-Instruct-v0.2')
llmchain=LLMChain(llm=model,prompt=prompt)
```

```python
from langchain.tools.retriever import create_retriever_tool

retriever_tool = create_retriever_tool(
    retriever=retriever,
    name="medical-search",
    description="Search for information about medicine related questions. For any questions about medicare, you must use this tool!",
)
```

```python
from langchain.prompts import PromptTemplate
from langchain_core.runnables.passthrough import RunnablePassthrough
prompt_template = """
### [INST] 
Instruction: Answer the question based on your 
knowledge about the context. Here is context to help:

{context}

### QUESTION:
{question} 

[/INST]
 """
prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=prompt_template,
)
```

```python
query = "what are the misuses of GenAI?" 

# retriever = db.as_retriever()

rag_chain = ( 
 {"context": retriever, "question": RunnablePassthrough()}
    | llmchain
)

output = rag_chain.invoke(query)
```

```python
from flask import Flask, request
from flask import render_template,jsonify


app = Flask(__name__)

@app.route('/airesponse',methods = ['POST'])
def get_rag_reponse():
    json_content = request.json
    query = json_content.get("query")
    rag_response = rag_chain.invoke(query)
    answer = rag_response['text'].split('\n')[-1].strip()
    answer_format = {'answer': answer}
    print('Successfull got rag response....')
    return answer_format

def start_app():
    app.run(debug=True)

if __name__=="__main__":
    start_app()
```
